**Regression**

The term regression is used when you try to find the relationship between variables.

In Machine Learning, and in statistical modeling, that relationship is used to predict the outcome of future events.

**Linear Regression**

Linear regression uses the relationship between the data-points to draw a straight line through all them.

This line can be used to predict future values.

In Machine Learning, predicting the future is very important.

Python has methods for finding a relationship between data-points and to draw a line of linear regression. We will show you how to use these methods instead of going through the mathematic formula.

In the example below, the x-axis represents age, and the y-axis represents speed. We have registered the age and speed of 13 cars as they were passing a tollbooth. Let us see if the data we collected could be used in a linear regression:

```python
import matplotlib.pyplot as plt
from scipy import stats
```

```python
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

plt.scatter(x, y)
plt.show()
```

```python
# drawing line of regression
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def myfunc(x):
  return slope * x + intercept

mymodel = list(map(myfunc, x))

plt.scatter(x, y)
plt.plot(x, mymodel)
plt.show()
print(std_err)
```

```python
print(slope)
print(intercept)
print(r)
print(p)
print(std_err)
```

Simple words mein ‚Äî **Linear Regression** ek aisi ML technique hai jo relationship dhoondhti hai do variables ke beech ‚Äî ek independent variable (x) aur ek dependent variable (y) ke darmiyan.

Soch le:

Tu check kar raha hai ke ‚Äúcar ka age‚Äù badhne se ‚Äúspeed‚Äù kam hoti hai ya nahi.

Age = x (independent variable)

Speed = y (dependent variable)

Goal ye hai ke hum ek seedhi line (straight line) fit karein jo sab data points ke beech se guzre maximum accuracy ke sath.

Equation hoti hai:

**ùë¶=ùëö√óùë•+ùëê**

m = **slope** (ye batata hai line kitni steep hai ‚Äî matlab x change hone se y kitna change hota hai)

c = **intercept** (jab x = 0 ho, tab y ki value kya hoti hai)

**stats.linregress(x, y)** ek shortcut hai jo automatically:

slope (m) calculate karta hai

intercept (c) nikalta hai

r (correlation coefficient) deta hai

p (p-value) deta hai

std_err (standard error) deta hai

Yani ek hi line mein poori regression analysis mil jati hai

**‚ÄúKya ye relationship jo humne find ki hai (x aur y ke beech) real hai, ya bas ek coincidence hai?‚Äù**

**p-value** isi ka jawab deti hai.
Ye batati hai ke result random chance se aaya hai ya genuinely ek relation hai.

üîπ Simple meaning:

Agar p chhoti ho (like < 0.05) ‚Üí to iska matlab hai relationship statistically significant hai.
Yani: x aur y ke beech real connection hai.

Agar p badi ho (> 0.05) ‚Üí to iska matlab hai random chance bhi ho sakta hai.

üß† Example:

Hum car age aur speed ka relation check kar rahe hain.
Agar p = 0.001 aata hai ‚Üí matlab strong evidence hai ke jaise age badhti hai, speed kam hoti hai.
Agar p = 0.6 aata hai ‚Üí matlab yeh pattern sirf random coincidence bhi ho sakta hai.

Desi example:
Tu kahay ‚ÄúMujhe lagta hai purani cars slow hoti hain.‚Äù
Agar p-value chhoti hai, matlab: ‚ÄúHaan bhai, data bhi yehi keh raha hai!‚Äù
Agar p-value badi hai: ‚ÄúBhai lagta to hai, par data itna confirm nahi hai.‚Äù

**Standard Error** basically batata hai ke:

**‚ÄúHamari fitted line (slope) kitni accurately estimate hui hai?‚Äù**

Ye ek uncertainty measure hai slope ke liye.

üîπ Simple meaning:

Agar std_err chhota hai ‚Üí slope accurate hai, model kaafi reliable hai.

Agar std_err bada hai ‚Üí slope mein uncertainty zyada hai (model weak ya noisy hai).

üß† Example:

Agar slope ka std_err = 0.5 hai ‚Üí line tight fit hui hai.
Agar std_err = 5 hai ‚Üí line random lag rahi hai, model unreliable hai.

Desi example:
Soch le tu car ka average speed guess kar raha hai purani aur nai car ke liye.
Agar tu har baar almost same value guess karta hai ‚Üí std_err low.
Agar tu kabhi 80 bolta hai, kabhi 120 ‚Üí std_err high.

Now we can use the information we have gathered to predict future values.

Example: **Let us try to predict the speed** of a 10 years old car.

To do so, we need the same myfunc() function from the example above:

```python
def predict(age):
    return slope * age + intercept

print(f"predicted speed for a 10 years old car is: {predict(10)}")
print(f"predicted speed for a 18 years old car is: {predict(18)}")
```

we can also see the grapth, we have plotted, to see that the values, predicted, are correct.

**Let us create an example where linear regression would not be the best method to predict future values.**

```python
x = [89,43,36,36,95,10,66,34,38,20,26,29,48,64,6,5,36,66,72,40]
y = [21,46,3,35,67,95,53,72,58,10,26,34,90,33,38,20,56,2,47,15]

slope, intercept, r, p, std_err = stats.linregress(x, y)

def prediction(x):
    return slope * x + intercept

myModel = list(map(prediction, x))

print(f"Slope: {slope}\nIntercept: {intercept}\nR-value: {r}\np-value: {p}\nStandard Error: {std_err}")

print(f"predicted value for x=80 is: {prediction(10)}")

plt.scatter(x, y)
plt.plot(x, myModel)
plt.show()
```

Linear regression uses the least square method.

The concept is to draw a line through all the plotted data points. The line is positioned in a way that it minimizes the distance to all of the data points.

The distance is called "residuals" or "errors".
